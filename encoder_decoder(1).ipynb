```python
from datasets import load_dataset
from transformers import AutoTokenizer
# Load dataset
dataset = load_dataset("gsarti/iwslt2017_context", "iwslt2017-ar-en")
```

```python
print(dataset)
print(dataset["train"][0])  # Inspect the first entry in the training set
```

```python
# Define the tokenizer here
tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
```

```python
# Tokenizer pour arabe et anglais
tokenizer_ar = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
tokenizer_en = AutoTokenizer.from_pretrained("bert-base-uncased")

def preprocess_data(batch):
    # Extract Arabic and English translations
    source = [item["ar"] for item in batch["translation"]]
    target = [item["en"] for item in batch["translation"]]

    # Tokenize the source and target texts
    source_encodings = tokenizer_ar(source, truncation=True, padding="max_length", max_length=64)
    target_encodings = tokenizer_en(target, truncation=True, padding="max_length", max_length=64)

    return {
        "input_ids": source_encodings["input_ids"],
        "attention_mask": source_encodings["attention_mask"],
        "labels": target_encodings["input_ids"],
    }

```

```python
processed_dataset = dataset.map(preprocess_data, batched=True)


```

```python
print(processed_dataset["train"][0])

```

```python
# Prepare PyTorch DataLoader
from torch.utils.data import DataLoader

def collate_fn(batch):
    return {
        "input_ids": torch.tensor([item["input_ids"] for item in batch]),
        "attention_mask": torch.tensor([item["attention_mask"] for item in batch]),
        "labels": torch.tensor([item["labels"] for item in batch]),
    }

train_loader = DataLoader(processed_dataset["train"], batch_size=32, collate_fn=collate_fn)
val_loader = DataLoader(processed_dataset["validation"], batch_size=32, collate_fn=collate_fn)
```

```python
# Define Encoder-Decoder Model (No Attention)
import torch.nn as nn
import torch

class Seq2SeqNoAttention(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.encoder = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.decoder = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, input_ids, attention_mask, labels=None):
        # Embedding and encoding
        embedded = self.embedding(input_ids)
        _, (hidden, cell) = self.encoder(embedded)

        # Prepare decoder input
        decoder_input = self.embedding(labels[:, :-1])  # Teacher forcing
        decoder_output, _ = self.decoder(decoder_input, (hidden, cell))

        # Final linear layer
        logits = self.fc(decoder_output)
        return logits
```

```python
# Instantiate Model
vocab_size = tokenizer.vocab_size
embedding_dim = 256
hidden_dim = 512
model = Seq2SeqNoAttention(vocab_size, embedding_dim, hidden_dim)
```

```python
# Define Loss and Optimizer
from torch.optim import AdamW
from torch.nn import CrossEntropyLoss

optimizer = AdamW(model.parameters(), lr=1e-4)
criterion = CrossEntropyLoss(ignore_index=tokenizer.pad_token_id)
```

```python
#Training Loop
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

num_epochs = 3
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    for batch in tqdm(train_loader):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        optimizer.zero_grad()
        logits = model(input_ids, attention_mask, labels)
        loss = criterion(logits.view(-1, vocab_size), labels[:, 1:].reshape(-1))
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {epoch_loss / len(train_loader)}")

# Save Model
torch.save(model.state_dict(), "seq2seq_no_attention.pth")
```

```python
import torch
from torch import nn

class Encoder(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)

    def forward(self, x):
        embedded = self.embedding(x)
        outputs, (hidden, cell) = self.rnn(embedded)
        return hidden, cell

class Decoder(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, x, hidden, cell):
        embedded = self.embedding(x).unsqueeze(1)  # Add time dimension
        outputs, (hidden, cell) = self.rnn(embedded, (hidden, cell))
        predictions = self.fc(outputs.squeeze(1))
        return predictions, hidden, cell

class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder

    def forward(self, src, trg):
        hidden, cell = self.encoder(src)
        outputs = []
        for t in range(trg.size(1)):
            output, hidden, cell = self.decoder(trg[:, t], hidden, cell)
            outputs.append(output)
        return torch.stack(outputs, dim=1)
```

```python
from torch.utils.data import DataLoader

# Define constants
INPUT_DIM = tokenizer.vocab_size
OUTPUT_DIM = tokenizer.vocab_size
EMBEDDING_DIM = 256
HIDDEN_DIM = 512
BATCH_SIZE = 32

# Initialize the model
encoder = Encoder(INPUT_DIM, EMBEDDING_DIM, HIDDEN_DIM)
decoder = Decoder(OUTPUT_DIM, EMBEDDING_DIM, HIDDEN_DIM)
model = Seq2Seq(encoder, decoder)

# Define optimizer and loss function
optimizer = torch.optim.Adam(model.parameters())
criterion = nn.CrossEntropyLoss()

# Prepare DataLoader
def collate_fn(batch):
    input_ids = torch.tensor([item["input_ids"] for item in batch])
    attention_mask = torch.tensor([item["attention_mask"] for item in batch])
    labels = torch.tensor([item["labels"] for item in batch])
    return input_ids, attention_mask, labels

train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, collate_fn=collate_fn)

# Training loop
for epoch in range(10):
    model.train()
    epoch_loss = 0
    for src, _, trg in train_loader:
        optimizer.zero_grad()
        output = model(src, trg[:, :-1])  # Exclude last token for teacher forcing
        loss = criterion(output.view(-1, OUTPUT_DIM), trg[:, 1:].reshape(-1))  # Shift target
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {epoch_loss/len(train_loader)}")
```
